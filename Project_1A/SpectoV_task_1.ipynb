Generate simple 2D datasets using make_blobs or make_moons from scikit-learn.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
```

```python
X, y = make_blobs(n_samples=300, centers=3, random_state=3)
plt.scatter(X[:, 0], X[:, 1])
plt.title("Blob")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Real clusters

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Blob - True Clusters")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Apply KMeans clustering first. Choose the number of clusters (k) and visualize the results using scatter plots with
color-coded groups

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=3,random_state=0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=100, c='red', label='Centroids')
plt.title("KMeans Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

DBSCAN

```python
from sklearn.cluster import DBSCAN


dbscan = DBSCAN(eps=0.3, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_dbscan)
plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

eps and min_sample for make_blobs

```python
X,Y=make_blobs(n_samples=300,centers=3,random_state=3)
eps_data=[0.1,0.5,1.0,2.0]
min_data=[3,5,10]

for m_data in min_data:
  plt.figure(figsize=(40,4))
  for e_data in eps_data:
    dbscan = DBSCAN(eps=e_data, min_samples=m_data)
    dbscan.fit(X)
    plt.subplot(1,4,eps_data.index(e_data)+1)
    plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_)
    plt.title("DBSCAN Clustering")
    plt.title(f"eps={e_data}, min={m_data}")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
  plt.show()
```

Trying for different random states

```python
values=[0,42,3]
for rs in values:
  X,Y=make_blobs(n_samples=300,centers=3,random_state=rs)
  plt.figure(figsize=(20,5))
  plt.subplot(1,4,1)
  plt.scatter(X[:,0],X[:,1])
  plt.title("Dataset")
  plt.xlabel("Feature 1")
  plt.ylabel("Feature 2")

  plt.subplot(1,4,2)
  plt.scatter(X[:,0],X[:,1],c=Y)
  plt.title("Orginal cluster")
  plt.xlabel("Feature 1")
  plt.ylabel("Feature 2")

  kmeans=KMeans(n_clusters=3,random_state=rs)
  kmeans.fit(X)
  plt.subplot(1,4,3)
  plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
  plt.title("KMeans Clustering")
  plt.xlabel("Feature 1")
  plt.ylabel("Feature 2")

  dbscan = DBSCAN(eps=0.3, min_samples=5)
  dbscan.fit(X)
  plt.subplot(1,4,4)
  plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_)
  plt.title("DBSCAN Clustering")
  plt.xlabel("Feature 1")
  plt.ylabel("Feature 2")

  plt.show()
```
