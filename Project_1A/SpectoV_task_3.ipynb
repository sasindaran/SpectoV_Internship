Understand and apply dimensionality reduction techniques like PCA (Principal Component Analysis) and t-SNE (t-distributed Stochastic Neighbor Embedding) to visualize high-dimensional data in 2D or 3D space. This helps in simplifying complex datasets and identifying patterns, which is important for organizing and interpreting sensor or image data in AR/VR systems.

```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt

iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

```python
print("Feature Names:")
print(iris.feature_names)
print("\nTarget Names:")
print(iris.target_names)
```

```python
print("First 5 rows:")
print(df.head())
```

```python
print(f"Dataset shape: {iris.data.shape}")
```

```python
print("Summary Statistics:")
print(df.describe())
```

Using PCA to convert 4D->2D

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(iris.data)

plt.figure()
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=iris.target, cmap='viridis')
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("PCA: Iris Dataset in 2D")

plt.show()

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

PCA reduces the dimensionality of data by combining correlated features into new uncorrelated variables so 3d,4d... all are possible use cases data compressing and visualization but in case of tsne only visualization

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, perplexity=30, learning_rate=1, init='random', random_state=42)
tsne_result = tsne.fit_transform(iris.data)

plt.figure()
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=iris.target, cmap='plasma')
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.title("t-SNE: Iris Dataset in 2D")

plt.show()
```

Each point represents a flower sample

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

axs[0].scatter(pca_result[:, 0], pca_result[:, 1], c=iris.target, cmap='viridis')
axs[0].set_title("PCA")
axs[0].set_xlabel("PC1")
axs[0].set_ylabel("PC2")

axs[1].scatter(tsne_result[:, 0], tsne_result[:, 1], c=iris.target, cmap='plasma')
axs[1].set_title("t-SNE")
axs[1].set_xlabel("t-SNE 1")
axs[1].set_ylabel("t-SNE 2")

plt.suptitle("PCA vs t-SNE on Iris Dataset", fontsize=16)
plt.tight_layout()
plt.show()
```

Tsne -cluster identification
pca-pattern discovery,feature reduction
